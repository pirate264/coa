```python
def identify_columns(df, name):
    print(f"\n--- Column Types in {name} ---")
    print("Categorical:", [col for col in df.select_dtypes('object')])
    print("Numerical:", [col for col in df.select_dtypes('number')])
```

```python

identify_columns(salary_df_cleaned, "Salary Data")
identify_columns(car_df_cleaned, "Used Car Data")
```

```python
import pandas as pd

df = pd.read_csv('Salary_Data.csv')

for col in df.select_dtypes(include='number'):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]

    print(f"\nColumn: {col}")
    print(f"Mean: {df[col].mean()}")
    print("Outliers using IQR method:\n", outliers if not outliers.empty else "None")
```

```python
from scipy.stats import zscore
import pandas as pd

# Load the data
df = pd.read_csv('Salary_Data.csv')

# Loop through numeric columns
for col in df.select_dtypes(include='number'):
    df[f'{col}_Z'] = zscore(df[col])
    outliers = df[abs(df[f'{col}_Z']) > 3]
    print(f"\nOutliers using Z-score method for '{col}':\n", outliers if not outliers.empty else "None")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('Salary_Data.csv')

# Select numeric columns
numeric_cols = df.select_dtypes(include='number').columns

# Set plot size
plt.figure(figsize=(12, len(numeric_cols) * 4))

# Create a boxplot for each numeric column
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(len(numeric_cols), 1, i)
    sns.boxplot(x=df[col], color='skyblue')
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```

```python
# Remove IQR-based outliers from Salary
def remove_iqr_outliers(salary_df_cleaned, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    return df[(df[column] >= lower) & (df[column] <= upper)]

df_no_outliers = remove_iqr_outliers(salary_df_cleaned, "Salary")
```

```python
print("Original Summary:\n", df["Salary"].describe())
print("\nCleaned Summary:\n", df_no_outliers["Salary"].describe())
```
