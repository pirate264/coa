```python
import pandas as pd
```

```python
vertebrate_data = {
    'Name': ['human', 'python', 'salmon', 'whale', 'frog', 'komodo', 'bat', 'pigeon', 
             'cat', 'leopard shark', 'turtle', 'penguin', 'porcupine', 'eel', 
             'salamander', 'qila monster', 'platvpus', 'owl', 'dolphin', 'eagle'],
    'Blood Type': ['warm', 'cold', 'cold', 'warm', 'cold', 'cold', 'warm', 'warm',
                  'warm', 'cold', 'cold', 'warm', 'warm', 'cold', 'cold', 'cold',
                  'warm', 'warm', 'warm', 'warm'],
    'Give Birth': ['yes', 'no', 'no', 'yes', 'no', 'no', 'yes', 'no', 'yes', 'yes',
                  'no', 'no', 'yes', 'no', 'no', 'no', 'no', 'no', 'yes', 'no'],
    'Can Fly': ['no', 'no', 'no', 'no', 'no', 'no', 'yes', 'yes', 'no', 'no',
               'no', 'no', 'no', 'no', 'no', 'no', 'no', 'yes', 'no', 'yes'],
    'Live in Water': ['no', 'no', 'yes', 'yes', 'sometimes', 'no', 'no', 'no', 'no',
                     'yes', 'sometimes', 'sometimes', 'no', 'yes', 'sometimes',
                     'no', 'no', 'no', 'yes', 'no']
}
```

```python
df = pd.DataFrame(vertebrate_data)
```

```python
print("Vertebrate Data:")
display(df)
```

```python
print(f"\nNumber of rows: {df.shape[0]}, Number of columns: {df.shape[1]}")
```

```python
print("\nUnique values in each column:")
for column in df.columns:
    print(f"\nColumn '{column}':")
    display(pd.DataFrame(df[column].value_counts()).rename(columns={column: 'Count'}))
```

```python
conditions = [
    # Rule 1: (Give Birth = no) ∧ (Can Fly = yes) → Birds
    (df['Give Birth'] == 'no') & (df['Can Fly'] == 'yes'),
    
    # Rule 2: (Give Birth = no) ∧ (Live in Water = yes) → Fishes
    (df['Give Birth'] == 'no') & (df['Live in Water'] == 'yes'),
    
    # Rule 3: (Give Birth = yes) ∧ (Blood Type = warm) → Mammals
    (df['Give Birth'] == 'yes') & (df['Blood Type'] == 'warm'),
    
    # Rule 4: (Give Birth = no) ∧ (Can Fly = no) → Reptiles
    (df['Give Birth'] == 'no') & (df['Can Fly'] == 'no'),
    
    # Rule 5: (Live in Water = sometimes) → Amphibians
    (df['Live in Water'] == 'sometimes')
]
```

```python
choices = ['Birds', 'Fishes', 'Mammals', 'Reptiles', 'Amphibians']
```

```python
import numpy as np 
```

```python
df['Class'] = np.select(conditions, choices, default='Unknown')
```

```python
print("\nClassified Vertebrate Data:")
display(df)
```

```python
print("\nDistribution of Classes:")
display(pd.DataFrame(df['Class'].value_counts()).rename(columns={'Class': 'Count'}))
```

```python
unclassified = df[df['Class'] == 'Unknown']
if not unclassified.empty:
    print("\nAnimals that couldn't be classified:")
    display(unclassified)
else:
    print("\nAll animals were successfully classified!")
```
